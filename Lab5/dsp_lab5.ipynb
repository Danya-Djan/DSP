```python
import numpy as np
from scipy import signal, linalg
from scipy.io import wavfile
import matplotlib.pyplot as plt
def DTFT(x, M):
    """
    Функция вычисляет значения ДВПФ в M точках на отрезке
    по оси нормированных частот [-0.5; 0.5]

    Parameters
    ----------
    x : complex
        входная последовательность отсчетов, первый отсчет при k=0
    M : int
        колличество точек на отрезке [-0.5; 0.5]

    Returns
    -------
    float
        значения оси нормированных частот,
        соотвествующих вычисленным значениям
        спектральной функции
    complex
        значения спектральной функции
    """
    return (np.fft.fftshift(np.fft.fftfreq(M, d=1)), np.fft.fftshift(np.fft.fft(x, M)))
```

1) Для каких оконных функций гармонические компоненты оказались различимы и почему?

Различимость гармоник зависит от ширины главного лепестка спектра оконной функции.
Чем уже главный лепесток, тем выше разрешение по частоте, и лучше различаются близкие гармоники.
Однако, уменьшение ширины главного лепестка обычно сопровождается увеличением боковых лепестков.
Прямоугольное окно имеет очень узкий главный лепесток, но высокие боковые лепестки → гармоники различимы, но с сильными помехами от боковых лепестков.
Окно Хэннинга, Хэмминга, Блэкмана: уменьшают уровень боковых лепестков, но делают главный лепесток шире → может быть сложнее различить близкие частоты.

2) Как влияет максимальный уровень боковых лепестков на результат спектрального анализа?

Высокий уровень боковых лепестков приводит к тому, что энергия сильных сигналов "растекается" по спектру.
Это маскирует слабые сигналы и создает ложные пики.
В случае двух близких гармоник высокая энергия одной гармоники может перекрывать вторую.
Чем ниже боковые лепестки, тем чище спектр: слабые сигналы не теряются на фоне "размазанных" сильных.
Однако за это платится ценой ширины главного лепестка и, следовательно, разрешающей способности.

---

**Алгоритм Герцеля**

1. Для $m=0,\,1,\, \dots,\, N-1$ определяем

$${{V}_{n}}[m]=x[m]+2\cos (2\pi n/N){{V}_{n}}[m-1]-{{V}_{n}}[m-2],\quad 0\le m\le N-1,$$

2. Вычисляем $X[n]$ (при $m=N-1$):

$$X[n]={{\left. {{y}_{n}}[m] \right|}_{m=N-1}}={{\left. \left( W_{N}^{-n}{{V}_{n}}[m]-{{V}_{n}}[m-1] \right) \right|}_{m=N-1}}.$$
$$V_n[-1]=V_n [-2] = 0$$

```python
N = 2048
fs, x = wavfile.read('xylophone9.wav')
x = x[32000:32000+N]
```

```python
plt.figure(figsize = (10, 4))
plt.plot(np.arange(N), abs(x))
```

```python
def w(n, k=1, N=N):
    return np.exp(-2j*np.pi*n*k/N)

for n in np.random.randint(low=0, high=N, size=10):
    # матричная форма
    W = linalg.dft(N)
    y1 = W @ x
    
    # герцель
    beta=np.array([w(n=-n, k=1, N=N), -1.0], dtype=complex)
    alpha=np.array([1.0, -2.0*np.cos(2*np.pi*n/N), 1.0], dtype=complex)
    y2 = signal.lfilter(beta, alpha, x)
    
    # быстрое преобразование фурье
    y3 = np.fft.fft(x)
    
    print(y1[n], y2[N-1], y3[n])
    assert(np.allclose(y1[n], y2[N-1]))
    assert(np.allclose(y1[n], y3[n]))
```

## Задача 4.2*. Метод Блэкмана-Тьюки для несмещенной оценки АКФ

Модифицируйте функции `auto_correlation` и `blackman_tukey_estimator` так, чтобы их можно было использовать и с несмещенной оценкой АКФ (со значениями соответствующих параметров `unbiased`). Постойте оценку СПМ по методу Блэкмана-Тьюки с несмещенной оценкой АКФ для данных из задачи 4.1. Выбор окна и размерность ДПФ осуществить самостоятельно. Сравните результат с задачей 4.1.

```python
def auto_correlation(x, N1, mode='biased'):
    if mode == 'biased':
        Rx = signal.correlate(x, x) / len(x)
        lag = np.arange(-(len(x)-1), len(x))
        return Rx[np.logical_and(lag<=N1-1, lag>=0)]
    else:
        Rx = signal.correlate(x, x) * 1./(x.size - abs(np.arange(-x.size+1, +x.size)))
        lag = np.arange(-(len(x)-1), len(x))
        return Rx[np.logical_and(lag<=N1-1, lag>=0)]

def blackman_tukey_estimator(x, window='bartlett', Nfft=None, auto_correlation_mode='biased', fs=1):
    N = len(x)
    N1 = int(np.ceil(N/10))
    if Nfft is None:
        Nfft=max(2**12, N)
    window = signal.get_window(window, 2*N1-1, fftbins=False)
    Rx = auto_correlation(x, N1, mode=auto_correlation_mode) * window[N1-1:2*N1:1]
    return fs*np.arange(Nfft)/Nfft, (2*np.fft.fft(Rx, Nfft).real - Rx[0])/fs
```

```python
plt.figure(figsize=(10, 4))
f, P_BT = blackman_tukey_estimator(x, window='hamming', Nfft=8*x.size, fs=fs, auto_correlation_mode='biased',)
plt.plot(f, P_BT, 'C0-', label = "blackman tukey, unbiased")

f, P_BT = blackman_tukey_estimator(x, window='hamming', Nfft=8*x.size, fs=fs, auto_correlation_mode='unbiased',)
plt.plot(f, P_BT, 'C1-', label = "blackman tukey, biased")

plt.legend(loc='upper right')
plt.xlim([0, fs/2])
plt.grid()
plt.title('blackman tukey')
plt.xlabel('$f$, Hz')
plt.ylabel('$G(f)$, Watt/Hz')
plt.tight_layout()
plt.show()
```
